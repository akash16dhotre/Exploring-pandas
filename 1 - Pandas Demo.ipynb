```python
country_group['SalaryUSD'].agg(['median','mean']).loc['Canada']
```

```python
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python')

#  This shows True for people who have worked with Python with other other languages; and false who is not working with Python
```

```python
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum() # Total people in India entered Python
```

```python
# We can apply a function and run it; We use lambda function
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
# Lambda is going to be a series and we pass in a function
```

```python
# We want to figure out a way to figure out what % of people in each country know python
country_respondant = df['Country'].value_counts()
country_respondant
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

Now we have to merge the above two

```python
# Let us concatenate
python_df = pd.concat([country_respondant, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df
```

Calculating Percentage using formula

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
python_df
```

Let us sort thhem with countries highly knowing python

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```

```python
python_df.head(50)
```
